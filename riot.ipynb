Get SummonerId and PUUID

```python
division = ['IV', 'III', 'II', 'I']
tier = ['IRON', 'BRONZE', 'SILVER', 'GOLD', 'PLATINUM', 'DIAMOND']

queue = 'RANKED_SOLO_5x5'
pageInt = 1

# Setup connection to Database
conn = psycopg2.connect(database="riotdb", user="postgres", password="password", host="localhost", port="5432")
cursor = conn.cursor()

for t in tier:
    for d in division:
        r = requests.get(f'https://na1.api.riotgames.com/lol/league/v4/entries/{queue}/{t}/{d}?page={pageInt}&api_key={API_KEY}').json()
        players = r
        time.sleep(120)
        for i in range(len(players)):
            # Add puuid to players JSON
            time.sleep(3)
            r = requests.get(f'https://na1.api.riotgames.com/lol/summoner/v4/summoners/{players[i]["summonerId"]}?api_key={API_KEY}')
            print(r)
            time.sleep(3)
            r = r.json()
            players[i]['puuid'] = r['puuid']
            insert_query = f'INSERT INTO summoners (puuid, summonerid, summonername, tier, rank, wins, losses, hotstreak) VALUES (%s, %s, %s, %s, %s, %s, %s, %s)'
            values = (players[i]['puuid'], players[i]['summonerId'], players[i]['summonerName'], players[i]['tier'], players[i]['rank'], players[i]['wins'], players[i]['losses'], players[i]['hotStreak'])
            
            try:
                # Execute INSERT Statement
                cursor.execute(insert_query, values)

                # Commit changes to Database
                conn.commit()
                
                print("Data inserted successfully.")
            except (Exception, psycopg2.Error) as error:
                # Roll back the transaction in case of an error:
                conn.rollback()
                print("Error inserting data:", error)

# Close cursor and db connection
cursor.close()
conn.close()
```
